```python
def canny_edge_detection(img: np.ndarray) -> np.ndarray:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 150, 300)

    return edges

def harris_corner_detection(img: np.ndarray) -> np.ndarray:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    dst = cv2.cornerHarris(gray, 3, 5, 0.1)
    corners = dst > 0.05 * dst.max()

    coord = np.argwhere(corners)
    
    return coord
```

```python
c1 = harris_corner_detection(img1)
c2 = harris_corner_detection(img2)

for y, x in c1:
    cv2.circle(depth1, (x,y), 3, (0,0,255), -1)

plt.imshow(depth1)
```
